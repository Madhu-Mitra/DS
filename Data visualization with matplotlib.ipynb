```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.arange(1,51)
y = np.random.normal(20,11,size=50)
y = np.square(y)
plt.figure(figsize=(16,6))
plt.plot(x,y, color="purple")
plt.fill_between(x, y,color="purple")
plt.show()
```

```python

x = np.arange(1,51)
y = np.random.normal(10,11,size=50)
y = np.square(y)
plt.figure(figsize=(16,6))

plt.fill_between( x, y, color="blue")  # #Changing Fill color
plt.plot(x, y, color='blue') # Color on edges
 
plt.title("$ Area $  $ chart $" , fontsize = 16)
plt.xlabel("$X$" , fontsize = 16)
plt.ylabel("$Y$" , fontsize = 16)


plt.show()
```

```python

x=np.arange(1,6)
y1 =  np.array([10,60,110,160,210])
y2 = np.array([11,70,120,170,220])
y3 = np.array([12,80,130,180,230])
y4 = np.array([40,90,140,190,240])
y5 = np.array([50,100,150,200,250])

plt.figure(figsize=(8,6))
plt.stackplot(x,y1,y2,y3,y4,y5, labels=['Y1','Y2','Y3','Y4','Y5'])
plt.legend(loc='upper left')
plt.show()
```
