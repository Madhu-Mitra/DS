```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.arange(1,51)
y = np.random.normal(20,11,size=50)
y = np.square(y)
plt.figure(figsize=(16,6))
plt.plot(x,y, color="purple")
plt.fill_between(x, y,color="purple")
plt.show()
```
